```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import  matplotlib.pyplot as plt 
from sklearn.naive_bayes import CategoricalNB,MultinomialNB 
from sklearn.preprocessing  import OrdinalEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
cust_df=pd.read_csv('categorical_data_cleaned.csv')
```

```python
cust_df.head()
```

```python
encoder=OrdinalEncoder()
scaler=MinMaxScaler()
```

```python
columns_encoded = ['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','Churn']
standard_scaled=['MonthlyCharges','TotalCharges']
```

```python
cust_df[columns_encoded]=encoder.fit_transform(cust_df[columns_encoded])
```

```python
cust_df[standard_scaled]=scaler.fit_transform(cust_df[standard_scaled])
```

```python
cust_df.head()
```

```python
cust_df.describe()
```

```python
X=cust_df.drop(columns=['Churn','PaperlessBilling','PaymentMethod'])
Y=cust_df['Churn']
```

```python
x_train , x_test ,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
NB_model=MultinomialNB()
```

```python
NB_model.fit(x_train,y_train)
```

```python
y_test_pred=NB_model.predict(x_test)
```

```python
print("Model Performance for Test Set ")
print("-Accuracy : {:.4f}".format(accuracy_score(y_test,y_test_pred)))
print("-F1-Score : {:.4f}".format(f1_score(y_test,y_test_pred,average='weighted')))
print("-Precision : {:.4f}".format(precision_score(y_test,y_test_pred)))
print("-Recall : {:.4f}".format(recall_score(y_test,y_test_pred)))
print("-ROC-AUC Score : {:.4f}".format(roc_auc_score(y_test,y_test_pred)))
```

```python
neural=MLPClassifier(activation ='relu', alpha= 0.0001, batch_size=128, hidden_layer_sizes= (100,100), learning_rate='adaptive',max_iter=500, solver ='sgd')
neural.fit(x_train,y_train)
```

```python
y_test_neural=neural.predict(x_test)
```

```python
print("Model Performance for Test Set ")
print("-Accuracy : {:.4f}".format(accuracy_score(y_test,y_test_neural)))
print("-F1-Score : {:.4f}".format(f1_score(y_test,y_test_neural,average='weighted')))
print("-Precision : {:.4f}".format(precision_score(y_test,y_test_neural)))
print("-Recall : {:.4f}".format(recall_score(y_test,y_test_neural)))
print("-ROC-AUC Score : {:.4f}".format(roc_auc_score(y_test,y_test_neural)))
```

```python
param_grid_ann = {
    'hidden_layer_sizes': [(50,)],  
    'activation': ['relu', 'tanh', 'logistic'],  
    'solver': ['adam', 'sgd'],  
    'alpha': [0.0001, 0.001, 0.01],  
    'learning_rate': ['constant', 'adaptive'], 
    'max_iter': [200, 300, 500],  
    'batch_size': [32, 64, 128],  
    # 'learning_rate_init': [0.001, 0.01, 0.1], 
    # 'early_stopping': [True, False],
    # ''(100,), (50, 50), (100, 100)''  
}
```

```python
GridSearchModels=[
    # ('Naive Bayes (Bernoullis)',BernoulliNB(),param_grid_bernoulli)
    ('MultiLayer Percepton',MLPClassifier(),param_grid_ann)   
]
```

```python
# Creating a Loop to Apply Hyperparamter Tunning to All the Specified Models  :

model_para = {}

for name,model,param in GridSearchModels :
    random=GridSearchCV(estimator=model,param_grid=param,cv=5,n_jobs=-1)

    random.fit(x_train,y_train)
    model_para[name] =random.best_params_

for model_name in model_para :
    print(f"************************* Best Parameters for {model_name} ********************************** ")
    print(model_para[model_name])
```
