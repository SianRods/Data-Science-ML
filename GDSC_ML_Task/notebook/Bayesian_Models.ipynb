```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import  matplotlib.pyplot as plt 
from sklearn.naive_bayes import CategoricalNB 
from sklearn.preprocessing  import OrdinalEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,roc_auc_score
from sklearn.linear_model import LogisticRegression

```

```python
cust_df=pd.read_csv('processed_data.csv')
```

```python
cust_df.head()
```

```python
# encoder=OrdinalEncoder()
scaler=MinMaxScaler()
```

```python
# columns_encoded = ['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','Churn']
standard_scaled=['MonthlyCharges','TotalCharges']
```

```python
# cust_df[columns_encoded]=encoder.fit_transform(cust_df[columns_encoded])
```

```python
cust_df[standard_scaled]=scaler.fit_transform(cust_df[standard_scaled])
```

```python
cust_df.head()
```

```python
X=cust_df.drop(columns=['Churn'])
Y=cust_df['Churn']
```

```python
x_train , x_test ,y_train,y_test =train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
NB_model=CategoricalNB()
```

```python
NB_model.fit(x_train,y_train)
```

```python
y_test_pred=NB_model.predict(x_test)
```

```python
print("Model Performance for Test Set ")
print("-Accuracy : {:.4f}".format(accuracy_score(y_test,y_test_pred)))
print("-F1-Score : {:.4f}".format(f1_score(y_test,y_test_pred,average='weighted')))
print("-Precision : {:.4f}".format(precision_score(y_test,y_test_pred)))
print("-Recall : {:.4f}".format(recall_score(y_test,y_test_pred)))
print("-ROC-AUC Score : {:.4f}".format(roc_auc_score(y_test,y_test_pred)))
```

```python
logistic_reg=LogisticRegression(penalty='l2',solver='liblinear')
```

```python
logistic_reg.fit(x_train,y_train)
```

```python
y_pred_logistic=logistic_reg.predict(x_test)
```

```python
print("Model Performance for Test Set ")
print("-Accuracy : {:.4f}".format(accuracy_score(y_test,y_pred_logistic)))
print("-F1-Score : {:.4f}".format(f1_score(y_test,y_pred_logistic,average='weighted')))
print("-Precision : {:.4f}".format(precision_score(y_test,y_pred_logistic)))
print("-Recall : {:.4f}".format(recall_score(y_test,y_pred_logistic)))
print("-ROC-AUC Score : {:.4f}".format(roc_auc_score(y_test,y_pred_logistic)))
```
